```python
#helper function for content cleaning
def text_clean(content):
    #lowercase the content
    content=content.lower()
    #remove punctuation
    content = re.sub('[()!?]', ' ', content)
    content = re.sub('\[.*?\]',' ', content)
    #remove non alphanumeric occurences
    content = re.sub("[^a-z0-9]"," ", content)
    #remove the @mention
    content = re.sub("@[A-Za-z0-9_]+","", content)
    #remove the hashtags
    content = re.sub("#[A-Za-z0-9_]+","", content)
    #remove any links
    content = re.sub(r"http\S+", "", content)
    content = re.sub(r"www.\S+", "", content)
    return content
```

```python
import random
import nltk
from nltk.corpus import wordnet

# Ensure you have the necessary NLTK resources
nltk.download('wordnet')

def get_synonyms(word):
    synonyms = wordnet.synsets(word)
    if not synonyms:
        return []
    return list(set([lemma.name() for synonym in synonyms for lemma in synonym.lemmas()]))

def synonym_replacement(sentence, n):
    words = sentence.split()
    new_words = words.copy()
    random_word_list = list(set([word for word in words if get_synonyms(word)]))
    random.shuffle(random_word_list)

    num_replaced = 0
    for random_word in random_word_list:
        synonyms = get_synonyms(random_word)
        if synonyms:
            synonym = random.choice(synonyms)
            new_words = [synonym if word == random_word else word for word in new_words]
            num_replaced += 1
        if num_replaced >= n:
            break

    return ' '.join(new_words)

def random_insertion(sentence, n):
    words = sentence.split()
    new_words = words.copy()
    for _ in range(n):
        if len(words) == 0:  # Check if there are words to choose from
            break

        synonym = get_synonyms(random.choice(words))
        if synonym:
            insert_pos = random.randint(0, len(new_words))  # Allow inserting at the end
            new_words.insert(insert_pos, random.choice(synonym))
    return ' '.join(new_words)

def random_swap(sentence, n):
    words = sentence.split()
    new_words = words.copy()
    if len(words) < 2:  # Ensure there are at least 2 words to swap
        return sentence

    for _ in range(n):
        idx1, idx2 = random.sample(range(len(words)), 2)
        new_words[idx1], new_words[idx2] = new_words[idx2], new_words[idx1]
    return ' '.join(new_words)

def random_deletion(sentence, p=0.3):
    words = sentence.split()
    if len(words) <= 1:  # If there's only one word, don't delete it
        return sentence

    # Remove words with probability p
    new_words = [word for word in words if random.uniform(0, 1) > p]

    # Check if new_words is empty; if so, return the original sentence or the first word
    if not new_words:
        return words[0]  # Return the first word to avoid empty sentence

    return ' '.join(new_words)

def eda(sentence, num_aug=4):
    augmented_sentences = []
    augmented_sentences.append(synonym_replacement(sentence, n=2))
    augmented_sentences.append(random_insertion(sentence, n=2))
    augmented_sentences.append(random_swap(sentence, n=2))
    augmented_sentences.append(random_deletion(sentence, p=0.3))
    return augmented_sentences
```

```python
# Augment the dataset
augmented_sentences = []
augmented_labels = []

for index, row in df.iterrows():
    sentence = row['content']
    label = row['intensity']

    # Generate augmented sentences for each row
    aug_sentences = eda(sentence, num_aug=4)

    # Append the original and augmented sentences along with labels
    augmented_sentences.append(sentence)  # Original sentence
    augmented_labels.append(label)  # Original label

    # Add augmented data
    augmented_sentences.extend(aug_sentences)
    augmented_labels.extend([label] * len(aug_sentences))

# Create a new DataFrame with augmented data
augmented_df = pd.DataFrame({
    'content': augmented_sentences,
    'intensity': augmented_labels
})

# Combine original and augmented data (optional, if you want to include the original data too)
final_df = pd.concat([df, augmented_df])

# Show the result
print(final_df)
```

```python
#Sentiment Classifier
#split the data into training and test set with balanced split based on target class
X_train, X_test, y_train, y_test = train_test_split(final_df["clean_content"],
                                                    final_df["intensity"],test_size=0.2,
                                                    stratify=final_df['intensity'])
```

```python
#check the shape of train and test data
print(X_train.shape[0],X_test.shape[0])
```

```python
#check the class balance in train data
print(y_train.value_counts())
```

```python
#check the target class balance in test data
print(y_test.value_counts())
```

```python
# Use the countvectoriser to convert text data into numerical/vector form
vectorizer = CountVectorizer(token_pattern=r'\b\w+\b')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
# train using Logistic Regression Model
alr1 = LogisticRegression(class_weight='balanced')
alr1.fit(X_train_vec,y_train)
print(confusion_matrix(alr1.predict(X_test_vec),y_test))
print(classification_report(alr1.predict(X_test_vec),y_test))
```

```python
from sklearn.model_selection import GridSearchCV

param_gridalr1 = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'newton-cg', 'saga']
}

grid_searchalr1 = GridSearchCV(alr1, param_gridalr1, cv=5, scoring='accuracy')
grid_searchalr1.fit(X_train_vec, y_train)

# Best model
best_alr1 = grid_searchalr1.best_estimator_
```

```python
#train and evaluate the RF classifier model
arf1 = RandomForestClassifier(n_estimators=100)
arf1.fit(X_train_vec,y_train)
print(confusion_matrix(arf1.predict(X_test_vec),y_test))
print(classification_report(arf1.predict(X_test_vec),y_test))
```

```python
#train and evaluate the Linear Support Vector Classifier model
aSVC1 = LinearSVC()
aSVC1.fit(X_train_vec,y_train)
print(confusion_matrix(aSVC1.predict(X_test_vec),y_test))
print(classification_report(aSVC1.predict(X_test_vec),y_test))
```

```python
# Define the parameter grid
param_grid_asvc1 = {
    'C': [0.01, 0.1, 1, 10, 100]
}

# Set up the grid search
grid_search_asvc1 = GridSearchCV(aSVC1, param_grid_asvc1, cv=5, scoring='accuracy')

# Fit the model
grid_search_asvc1.fit(X_train_vec, y_train)

# Get the best model and parameters
best_asvc1 = grid_search_asvc1.best_estimator_
print(f"Best parameters: {grid_search_asvc1.best_params_}")
print(f"Best cross-validation accuracy: {grid_search_asvc1.best_score_}")

```
